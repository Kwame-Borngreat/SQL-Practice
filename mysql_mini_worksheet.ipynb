```python
%%sql
SELECT
    type_of_water_source,
    COUNT(type_of_water_source) AS total_people_served
FROM 
    water_source
GROUP BY type_of_water_source
ORDER BY total_people_served DESC;
```

```python
%%sql
SELECT
    type_of_water_source,
    ROUND(AVG(number_of_people_served)) AS aver_num_people_served
    
FROM 
    water_source
GROUP BY type_of_water_source
```

```python
%%sql
SELECT
    type_of_water_source,
    SUM(number_of_people_served) AS total_people_per_source
    
FROM 
    water_source
GROUP BY type_of_water_source
```

```python
%%sql
SELECT
    type_of_water_source,
    SUM(number_of_people_served) AS total_people_per_source, 
    round(SUM(number_of_people_served)/
          (SELECT SUM(number_of_people_served) FROM water_source) *100) AS Share_Percentage
FROM 
    water_source
GROUP BY
    type_of_water_source
ORDER BY Share_Percentage DESC;
```

# SOLUTION



```python
%%sql
SELECT
    type_of_water_source,
    SUM(number_of_people_served) AS total_people_per_source,
RANK () OVER(
    ORDER BY SUM(number_of_people_served) DESC
    ) AS Rank_by_total_people_per_source
FROM 
    water_source
WHERE 
    type_of_water_source<>"tap_in_home"
GROUP BY type_of_water_source
```

```python
%%sql
SELECT
    source_id,
    type_of_water_source,
    (number_of_people_served), 
RANK() OVER(
    PARTITION by type_of_water_source
    ORDER BY(number_of_people_served) DESC
    ) AS priority_rank
FROM 
    water_source
WHERE 
    type_of_water_source<>"tap_in_home"
LIMIT 10;
```

# ANALYSING QUEUES

```python

%%sql

SELECT 
    *

FROM visits
LIMIT 10;
```

```python
%%sql
SELECT
    SUBSTRING( MIN(time_of_record),1,10) AS Survey_Starts, 
    SUBSTRING(Max(time_of_record),1,10)  AS Survey_Ends,
    DATEDIFF(Max(time_of_record),MIN(time_of_record))

FROM visits

LIMIT 10;
```

```python
#AVERAGE QUEUE TIME
```

```python
%%sql
SELECT
AVG(time_in_queue) AS avg_queue_time_in_minutes
FROM visits
WHERE time_in_queue<>0
```

```python
%%sql
SELECT
    AVG(NULLIF(time_in_queue,0)) AS avg_queue_time_in_minutes
FROM visits
```

```python
%%sql
SELECT 
    DAYNAME(time_of_record) AS Day_of_week,
    round(AVG(time_in_queue)) AS average_queue_time
    
FROM 
    visits
GROUP BY DAYNAME(time_of_record)
ORDER BY round(AVG(time_in_queue)) DESC
LIMIT 10;
```

```python
%%sql
SELECT 
    TIME_FORMAT(TIME(time_of_record),"%H:00" )AS time_of_the_day,
    round(AVG(time_in_queue)) AS average_queue_time
    
FROM 
    visits
WHERE time_in_queue <>0
GROUP BY TIME_FORMAT(TIME(time_of_record),"%H:00" )
ORDER BY time_of_the_day
LIMIT 24;
```
